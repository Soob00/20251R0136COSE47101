## 범주형 변수 : 히스토그램 사용해서 질병 빈도수 시각화

어떤 범주에서 정신질병 수가 늘어나는지 같은 걸 알 수 있을듯
그냥 결측치 단순하게 제거한 데이터 사용했음.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv('/content/결측치가 포함된 모든 행 제거.csv')
```

```python
print(df.columns)
```

```python
def plot_normalized_histogram(var, value_col='NUMMHS', missing_val=-9):

    # 비율 계산
    prop_df = (
        df.groupby([var, value_col])
        .size()
        .groupby(level=0)
        .transform(lambda x: x / x.sum())
        .reset_index(name='proportion')
    )

    # 시각화
    g = sns.catplot(
        data=prop_df, x=value_col, y='proportion', col=var,
        kind='bar', col_wrap=4, height=4, sharey=True
    )

    g.set_titles(col_template=f"{var} = {{col_name}}")
    g.set_axis_labels(value_col, "Proportion")
    g.fig.suptitle(f"{value_col} 비율 분포 by {var}", y=1.05)
    plt.tight_layout()
    plt.show()
```

```python
variables = ['AGE','GENDER', 'EDUC', 'EMPLOY', 'MARSTAT', 'RACE', 'ETHNIC','LIVARAG','IJSSERVICE','RTCSERVICE','OPISERVICE','SPHSERVICE']
```

```python
for var in variables:
    plot_normalized_histogram(var)
```

### 결과 해석

단순하게 숫자 세는 거는 그냥 인원 수 많은 카테고리가 크게 나오는거 때문에 비율로 시각화를 해보았어요 분포가 궁금한거니까
... 여기 있는 것들은 단순하게 봐선 큰 차이가 없어보임....

```python
def plot_group_mean(var, value_col='NUMMHS', missing_val=-9):
    df_valid = df[df[var] != missing_val]
    plt.figure(figsize=(10, 5))
    sns.barplot(data=df_valid, x=var, y=value_col, ci='sd')
    plt.title(f'{value_col} 평균 by {var}')
    plt.ylabel('평균 진단 수')
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_group_mean('EMPLOY')
plot_group_mean('EDUC')
plot_group_mean('GENDER')
plot_group_mean('ETHNIC')
```

그레서 몇개 범주 별 평균 진단 병 수 보는데 이걸로만 봐서는 잘 모르겠음

## 상관분석

### ordinal  Spearman 이용

### 범주형 변수 간 상관분석 : cramers'v 사용

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k)-1)))

def cramers_v_matrix(df, cat_cols):
    result = pd.DataFrame(index=cat_cols, columns=cat_cols)
    for col1 in cat_cols:
        for col2 in cat_cols:
            try:
                result.loc[col1, col2] = cramers_v(df[col1], df[col2])
            except:
                result.loc[col1, col2] = np.nan
    return result.astype(float)
```

```python
cat_vars = ['GENDER', 'RACE', 'ETHNIC', 'EMPLOY', 'EDUC', 'MARSTAT',
            'REGION', 'DIVISION', 'CMPSERVICE', 'SMISED', 'SAP', 'VETERAN', 'LIVARAG']

cramers_matrix = cramers_v_matrix(df, cat_vars)
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(cramers_matrix, annot=True, fmt=".2f", cmap='YlGnBu')
plt.title("(Cramér’s V)")
plt.tight_layout()
plt.show()
```

region, division은 모두 지역관련 칼럼이라 서로 포함되는 관계기 때문에 1이 나온 듯 합니다.

눈에 뛸 정도로 상관관계가 보이는 것은 region, division이랑  SAP인데 미국이 지역별로 약물관련 문제가 있어서 그거때문인 것 같아요
