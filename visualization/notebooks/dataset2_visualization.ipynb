## 범주형 변수 : 히스토그램 사용해서 질병 빈도수 시각화

어떤 범주에서 정신질병 수가 늘어나는지 같은 걸 알 수 있을듯
그냥 결측치 단순하게 제거한 데이터 사용했음.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/content/결측치가 포함된 모든 행 제거.csv')
```

```python
print(df.columns)
```

```python
def plot_normalized_histogram(var, value_col='NUMMHS', missing_val=-9):

    # 비율 계산
    prop_df = (
        df.groupby([var, value_col])
        .size()
        .groupby(level=0)
        .transform(lambda x: x / x.sum())
        .reset_index(name='proportion')
    )

    # 시각화
    g = sns.catplot(
        data=prop_df, x=value_col, y='proportion', col=var,
        kind='bar', col_wrap=4, height=4, sharey=True
    )

    g.set_titles(col_template=f"{var} = {{col_name}}")
    g.set_axis_labels(value_col, "Proportion")
    g.fig.suptitle(f"{value_col} 비율 분포 by {var}", y=1.05)
    plt.tight_layout()
    plt.show()
```

```python
variables = ['AGE','GENDER', 'EDUC', 'EMPLOY', 'MARSTAT', 'RACE', 'ETHNIC','LIVARAG','IJSSERVICE','RTCSERVICE','OPISERVICE','SPHSERVICE']
```

```python
for var in variables:
    plot_normalized_histogram(var)
```

### 결과 해석

단순하게 숫자 세는 거는 그냥 인원 수 많은 카테고리가 크게 나오는거 때문에 비율로 시각화를 해보았어요 분포가 궁금한거니까
... 여기 있는 것들은 단순하게 봐선 큰 차이가 없어보임....

```python
def plot_group_mean(var, value_col='NUMMHS', missing_val=-9):
    df_valid = df[df[var] != missing_val]
    plt.figure(figsize=(10, 5))
    sns.barplot(data=df_valid, x=var, y=value_col, ci='sd')
    plt.title(f'{value_col} 평균 by {var}')
    plt.ylabel('평균 진단 수')
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_group_mean('EMPLOY')
plot_group_mean('EDUC')
plot_group_mean('GENDER')
plot_group_mean('ETHNIC')
```

그레서 몇개 범주 별 평균 진단 병 수 보는데 이걸로만 봐서는 잘 모르겠음
barplot 사용해본건데
